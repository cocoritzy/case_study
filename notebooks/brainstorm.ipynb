```python
## do fields name change according seasons ?

# Let's count how many unique seasons exist per field name
field_season_counts = df_season.groupby('farm_season_field_name')['season'].nunique().reset_index()
field_season_counts.columns = ['field_name', 'num_seasons']

# Sort to see which farms appear in multiple seasons
field_season_counts = field_season_counts.sort_values(by='num_seasons', ascending=False)

import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-white')  # clean background
bar_color = "#32403D"  # Soil Capital muted blue

plt.figure(figsize=(12, 6))
bars = plt.bar(field_season_counts['field_name'], field_season_counts['num_seasons'], color=bar_color)

plt.title('Number of Seasons per Field', fontsize=14, fontname='DejaVu Sans')
plt.xlabel('Field Name', fontname='DejaVu Sans', fontsize=10)
plt.ylabel('Number of Seasons', fontname='DejaVu Sans', fontsize=10)
plt.xticks(rotation=45, ha='right', fontsize=9, fontname='DejaVu Sans')
plt.yticks(fontsize=9, fontname='DejaVu Sans')

```

```python
## how does regenerative agriculture practices evolve over the years ?
# List of boolean columns you're interested in
bool_cols = [
    'has_cover_crops', 'organic', 'irrigated',
    'artificial_drainage', 'agroforestry', 'was_pasture'
]

# Group by season and compute mean (mean of bools = proportion True)
practice_by_year = df_season.groupby('season')[bool_cols].mean().reset_index()

# Multiply by 100 to get %
practice_by_year[bool_cols] = practice_by_year[bool_cols] * 100

practice_by_year[bool_cols]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Reuse your DataFrame: practice_by_year
# Columns: season + has_cover_crops, organic, ...

labels = practice_by_year['season']
bool_cols = ['has_cover_crops', 'organic', 'irrigated', 'artificial_drainage', 'agroforestry', 'was_pasture']

x = np.arange(len(labels))  # label locations
width = 0.12  # width of each bar
offsets = np.linspace(-width*len(bool_cols)/2, width*len(bool_cols)/2, len(bool_cols))

plt.figure(figsize=(14, 6))

for i, col in enumerate(bool_cols):
    plt.bar(x + offsets[i], practice_by_year[col], width, label=col.replace('_', ' ').title())

plt.xticks(x, labels, rotation=45)
plt.ylabel('Percentage of Fields (%)')
plt.title('Adoption of Regenerative Practices by Season')
plt.legend(title='Practice', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your Excel file
df = pd.read_excel("prod_views_data_view_farm_season_field.xlsx")

# Convert soil pH to numeric midpoints
def ph_to_numeric(value):
    if isinstance(value, str):
        value = value.strip()
        if value == "< 5.5":
            return 5.0
        elif value == "5.5 - 7.3":
            return (5.5 + 7.3) / 2  # 6.4
        elif value == "7.3 - 8.5":
            return (7.3 + 8.5) / 2  # 7.9
    return None

df['soil_ph'] = df['soil_ph'].apply(ph_to_numeric)

# Compute averages per programme year
yearly_summary = df.groupby('programme_year')[
    ['yield', 'organic_matter', 'soil_ph', 'has_cover_crops',
     'irrigated', 'with_companion_crop', 'em_ghg_total', 'em_total',
     'em_fert_production', 'em_fuel']
].mean().reset_index()

# Compute crop diversity (number of unique crop types per year)
crop_diversity = df.groupby('programme_year')['crop_name'].nunique().reset_index()
crop_diversity.columns = ['programme_year', 'num_unique_crops']

# Merge it into the main summary
yearly_summary = pd.merge(yearly_summary, crop_diversity, on='programme_year')

# Columns to plot (excluding programme_year)
columns_to_plot = yearly_summary.columns[1:]

# Create grid of bar charts
n_cols = 3
n_rows = (len(columns_to_plot) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    axes[i].bar(yearly_summary['programme_year'], yearly_summary[col] *100, color='steelblue')
    axes[i].set_title(col.replace('_', ' ').title())
    axes[i].set_xlabel("Programme Year")
    axes[i].set_xticks(yearly_summary['programme_year'].astype(int))
    axes[i].set_ylabel("Average Value (%)")
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle("Farm Resilience Indicators per Programme Year", fontsize=16, y=1.02)
plt.show()
```

```python
# Group by year and field
field_year_summary = df.groupby(['programme_year', 'farm_season_field_name'])[
    ['yield', 'organic_matter', 'soil_ph', 'has_cover_crops',
     'irrigated', 'with_companion_crop', 'em_ghg_total', 'em_total',
     'em_fert_production', 'em_fuel']
].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

# Pick the metric you want to plot per field
metric = 'yield'

# Get unique field names
fields = field_year_summary['farm_season_field_name'].unique()

# Set up grid
n_cols = 3
n_rows = (len(fields) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 4 * n_rows))
axes = axes.flatten()

# Plot for each field
for i, field in enumerate(fields):
    data = field_year_summary[field_year_summary['farm_season_field_name'] == field]
    axes[i].bar(data['programme_year'], data[metric], color='seagreen')
    axes[i].set_title(f"{field} — {metric.title()}")
    axes[i].set_xlabel("Programme Year")
    axes[i].set_ylabel("Value")
    axes[i].set_xticks(data['programme_year'].astype(int))
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle(f"{metric.title()} per Field over Time", fontsize=16, y=1.02)
plt.show()
```

```python
## merge

# Merge shapefile and metrics on the field name
gdf_merged = gdf.merge(field_year_summary, left_on='farm_sea_2', right_on='farm_season_field_name', how='left')

```

```python
import matplotlib.pyplot as plt

# Select your metric and the years to show
metric = 'yield'
years = sorted(gdf_merged['programme_year'].dropna().unique())[:4]  # Limit to first 4 years

# Set up the 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes = axes.flatten()

# Plot each year's data
for i, year in enumerate(years):
    data = gdf_merged[gdf_merged['programme_year'] == year]

    data.plot(
        column=metric,
        cmap='YlGnBu',
        edgecolor='black',
        linewidth=0.5,
        legend=True,
        ax=axes[i],
        missing_kwds={"color": "lightgrey", "label": "No data"}
    )

    axes[i].set_title(f"{metric.title()} by Field — Year {year}", fontsize=12)
    axes[i].axis('off')

# Remove unused subplot if less than 4 years
for j in range(i + 1, 4):
    fig.delaxes(axes[j])

plt.suptitle(f"{metric.title()} Across Fields (2x2 Grid)", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the metrics you want to visualize
metrics = ['yield', 'organic_matter', 'soil_ph', 'has_cover_crops',
           'irrigated', 'with_companion_crop', 'em_ghg_total', 'em_total',
           'em_fert_production', 'em_fuel']

# Limit to the first 4 years available
years = sorted(gdf_merged['programme_year'].dropna().unique())[:4]

# Loop through each metric
for metric in metrics:
    fig, axes = plt.subplots(2, 2, figsize=(6, 5))
    axes = axes.flatten()

    for i, year in enumerate(years):
        data = gdf_merged[gdf_merged['programme_year'] == year]

        # Plot the map for each year
        data.plot(
            column=metric,
            cmap='YlGnBu',
            edgecolor='black',
            linewidth=0.5,
            legend=True,
            ax=axes[i],
            missing_kwds={"color": "lightgrey", "label": "No data"}
        )

        axes[i].set_title(f"{metric.replace('_', ' ').title()} — Year {year}", fontsize=12)
        axes[i].axis('off')

    # Hide unused axes if < 4 years
    for j in range(i + 1, 4):
        fig.delaxes(axes[j])

    # Add super title and show the whole grid
    plt.suptitle(f"{metric.replace('_', ' ').title()} Across Fields (2×2 Grid)", fontsize=16, y=1.02)
    plt.tight_layout()
    plt.show()
```
