```python
plt.figure(figsize=(6, 3))
sns.histplot(df['quantity'].dropna(), bins=30, color='salmon')
plt.title("Distribution of Active Pesticide Applied (kg)")
plt.xlabel("Active Substance (kg)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#top = df.groupby('farm_season_field_id')['active_kg'].sum().sort_values(ascending=False).head(10)
top = df.groupby('pesticide_name')['quantity'].sum().sort_values(ascending=False).head(10)
top.plot(kind='barh', color='salmon')
plt.xlabel("Total Active Ingredient Applied (kg)")
plt.title("Top 10 Most Used Pesticides")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Compute bounds
q = df['quantity'].dropna()
lower = q.quantile(0.02)
upper = q.quantile(0.98)

# Count how many are below or above the bounds
outliers = q[(q < lower) | (q > upper)]

print(outliers)
```

## aggregate per season and field

```python

# Group by year and calculate total & average for season
yearly_stats = df.groupby("season")["quantity"].agg(["sum", "mean", "count"]).reset_index()

# Group by year and calculate total & average for season and field
field_stats = df.groupby("farm_season_field_id")["quantity"].agg(["sum", "mean", "count"]).reset_index()
field_stats
```

```python
df['farm_season_field_id'].nunique()
```

```python
df_sum= df.groupby('farm_season_field_id')['quantity'].sum()

df_sum.index.name = "farm_season_field_id"
df_sum.rename("pesticide_in_kg", inplace=True)
```

```python
# Sort by quantity (optional for cleaner visualization)
df_sum_sorted = df_sum.sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 4))
df_sum_sorted.plot(kind='bar', color='skyblue')
plt.title('Total Quantity per Field')
plt.xlabel('Field ID')
plt.ylabel('Total Quantity')
plt.tight_layout()
plt.show()
```
